```python
import re
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

# Data input

```python
df= pd.read_csv('../input/sms-spam-collection-dataset/spam.csv')
df.head()
```

```python
len(df)
```

# Text Preprocessing and Corpus creation
- Removing special characters & Punctuations
- Lowering case
- Split into words
- Lemmatize the text 

```python
wn=WordNetLemmatizer()
corpus=[]

#ps = PorterStemmer()
for i in range(0,len(df)):
    text = re.sub('[^a-zA-Z]', ' ', df['v2'][i])
    text = text.lower()
    text = text.split()
    text = [wn.lemmatize(word) for word in text if not word in stopwords.words('english')]
    text = [c for c in text if c not in string.punctuation]
    
    text = ' '.join(text)
    corpus.append(text)
corpus[:5]
```

# Setting up Target(label) variable

```python
y=pd.get_dummies(df['v1'])
y=y['spam']
```

# Convert to TF-IDF vector & use Naive bayes classifier for prediction

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=2000)
X = cv.fit_transform(corpus).toarray()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
# Naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)

pred=spam_detect_model.predict(X_test)
```

# Calculate accuracy

```python
cm= confusion_matrix(y_test, pred)
acc= accuracy_score(y_test, pred)
cm, acc
```
